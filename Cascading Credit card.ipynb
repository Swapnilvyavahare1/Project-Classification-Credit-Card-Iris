```python
# Taking 10% dataset for building and predicting model
# Take all data together and then perform afterwards
```

```python
df1= df.sample(frac=0.1, random_state=1)
df1.shape
```

```python
df1['Class'].value_counts()
```

```python
fraud1= df1[df1['Class']==1]
normal1= df1[df1['Class']==0]

print(fraud1.shape, normal1.shape)
```

```python
df1.columns
```

```python
columns= df1.columns.tolist()
columns
```

```python
# Split the data into x and y
columns= [c for c in columns if c not in ['Class']]
target= 'Class'
```

```python
columns
```

```python
target
```

```python
x= df1[columns]
y= df1[target]

state= np.random.RandomState(101)
x_outlier= state.uniform(low=0, high=1, size=(x.shape[0], x.shape[1]))
print(x.shape, y.shape)
```

```python
x.head()
```

```python
# Split the data in train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.7, random_state=101)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_train_lr= lr.predict(x_train)
y_pred_test_lr= lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(classification_report(y_train, y_pred_train_lr))
print('..........................................')
print(classification_report(y_test, y_pred_test_lr))
print('..........................................')
print(confusion_matrix(y_train, y_pred_train_lr))
print('..........................................')
print(confusion_matrix(y_test, y_pred_test_lr))
print('..........................................')
print(accuracy_score(y_train, y_pred_train_lr))
print('..........................................')
print(accuracy_score(y_test, y_pred_test_lr))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train, y_train)
```

```python
y_pred_train_dt= dt.predict(x_train)
y_pred_test_dt= dt.predict(x_test)

print(classification_report(y_train, y_pred_train_dt))
print('..........................................')
print(classification_report(y_test, y_pred_test_dt))
print('..........................................')
print(confusion_matrix(y_train, y_pred_train_dt))
print('..........................................')
print(confusion_matrix(y_test, y_pred_test_dt))
print('..........................................')
print(accuracy_score(y_train, y_pred_train_dt))
print('..........................................')
print(accuracy_score(y_test, y_pred_test_dt))
```
